```python
melt_df=df.melt(id_vars= df.columns[0],value_vars=df.columns[1:],var_name='Movie',value_name='rating')


melt_df.fillna(0,inplace=True)
melt_df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics.pairwise import cosine_similarity
from scipy import sparse
```

```python

from scipy.sparse import csr_matrix
movies_users=melt_df.pivot(index='Movie', columns='user_id', values='rating')
#movies_users=df_final.pivot()
movies_users

```

```python

from fuzzywuzzy import process
def recommender(movie_name, data,model, n_recommendations ):
    model.fit(data)
    idx=process.extractOne(movie_name, df_dummy1['Movie_id'])[2]
    print(idx)
    print('Movie Selected: ',df_dummy1['Movie_id'][idx], 'Index: ',idx)
    print('Searching for recommendations.....')
    distances, indices=model.kneighbors(data[idx], n_neighbors=n_recommendations)
    print(distances,indices)
    for i in indices:
        print(df_dummy1['Movie_id'][i])
    
recommender('Movie18', mat_movies_users, model_knn,5)
```
